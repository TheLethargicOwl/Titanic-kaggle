```python
import numpy as np
import pandas as pd
from sklearn import tree,linear_model,neural_network
import re
```

```python
train = pd.read_csv("train.csv",sep = ",")
train.isnull().sum()
train.head()
```

```python
train.drop(["PassengerId", "Ticket","Cabin"],axis = 1,inplace = True)
```

```python
#train.Cabin[train.Cabin.isna()==0]=0
#train.Cabin[train.Cabin.isna()==1]=1
```

```python
Sex = pd.get_dummies(train.Sex)
Sex = Sex.female
train.Sex = Sex
```

```python
Embarked = pd.get_dummies(train.Embarked)
Embarked = Embarked.iloc[:,0:2]
train = pd.concat([train,Embarked],axis = 1)
train.drop(['Embarked'],axis = 1,inplace = True)
```

```python
#train_age = train[train.Age.isna()==0]
#test_age = train[train.Age.isna()==1]
```

```python
lis=[]
for i in train.Name:
    start = re.escape(", ")
    end   = re.escape(". ")
    result = re.search('%s(.*)%s' % (start, end), i).group(1)
    lis.append(result)
for i in range(len(train)):
    train.Name[i]=lis[i]
for i in range(len(train)):
    if train.Name[i]!='Master' and train.Name[i]!='Miss' and train.Name[i]!='Mr' and train.Name[i]!='Mrs':
        train.Name[i]='Mr'
```

```python
train['Sum']=0
for i in range(len(train)):
    train.Sum[i] = train.SibSp[i]+train.Parch[i]
```

```python
#train.drop(['SibSp','Parch'],axis =1,inplace = True)
```

```python
for i in range(len(train)):
    if(train.Age.isnull()[i]==True):
        if(train.Name[i]=='Master'):
            train.Age[i]=train.groupby(['Name']).Age.mean()['Master']
        if(train.Name[i]=='Miss'):
             train.Age[i]=train.groupby(['Name']).Age.mean()['Miss']
        if(train.Name[i]=='Mr'):
            train.Age[i]=train.groupby(['Name']).Age.mean()['Mr']
        if(train.Name[i]=='Mrs'):
            train.Age[i]=train.groupby(['Name']).Age.mean()['Mrs']
```

```python
for i in range(len(train)):
    if train.Name[i]=='Mr':
        train.Name[i]=1
    if train.Name[i]=='Mrs':
        train.Name[i]=2
    if train.Name[i]=='Miss':
        train.Name[i]=3
    if train.Name[i]=='Master':
        train.Name[i]=4
```
